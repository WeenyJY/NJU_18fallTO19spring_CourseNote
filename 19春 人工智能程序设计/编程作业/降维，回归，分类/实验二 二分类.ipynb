```python
import pandas as pd
import numpy as np
columns=['number','result']+['data'+str(x) for x in range(1,31)]
cancer=pd.read_csv('C:/Users/Jaqen/Desktop/wdbc.csv',names=columns)
cancer
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X=cancer.iloc[:,2:].values
Y=cancer.result.values
pca=PCA(n_components=2)
X_=pca.fit_transform(X) #é™ç»´
print(pca.explained_variance_ratio_) #æŸ¥çœ‹ä¸»æˆåˆ†æ¯”ä¾‹

f1=plt.figure()
ax=f1.add_subplot(111)
X0=np.hstack([Y.reshape(-1,1),X_])
a=np.array([x for x in X0 if x[0]=='M'])
b=np.array([x for x in X0 if x[0]=='B'])
ax.scatter(a[:,1],a[:,2],marker='o',c='g')
ax.scatter(b[:,1],b[:,2],marker='*',c='m')
plt.show()       #æ˜¾ç„¶X_åˆ†æˆäº†ä¸¤ç°‡ï¼Œå¯ä»¥äºŒåˆ†ç±»
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

def I(b):
    if b==True:
        return 1
    else:
        return 0
def recall_rate(predict,primary):   #å¬å›žçŽ‡ä¸ºè¯„ä»·æŒ‡æ ‡,å¬å›žçŽ‡è¶Šå¤§ï¼Œæ¨¡åž‹æ•ˆæžœè¶Šå¥½
    count=0
    countf=0
    for i in range(predict.shape[0]):
        count+=I(predict[i]==primary[i]=='M')
        countf+=I(predict[i]=='B' and primary[i]=='M')
    return count/(count+countf)

model=LogisticRegression(multi_class='ovr')
X_train, X_test, Y_train, Y_test = train_test_split(X_, Y, test_size=0.3, random_state=42)
r=model.fit(X_train,Y_train)
y_train=r.predict(X_train)  #è®­ç»ƒé›†çš„é¢„æµ‹å€¼
y_test=r.predict(X_test)   #æµ‹è¯•é›†çš„é¢„æµ‹å€¼

print('è®­ç»ƒè¯¯å·®:',recall_rate(y_train,Y_train))
print('æ³›åŒ–è¯¯å·®:',recall_rate(y_test,Y_test))
```
