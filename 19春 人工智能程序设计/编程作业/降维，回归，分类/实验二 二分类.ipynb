```python
import pandas as pd
import numpy as np
columns=['number','result']+['data'+str(x) for x in range(1,31)]
cancer=pd.read_csv('C:/Users/Jaqen/Desktop/wdbc.csv',names=columns)
cancer
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

X=cancer.iloc[:,2:].values
Y=cancer.result.values
pca=PCA(n_components=2)
X_=pca.fit_transform(X) #降维
print(pca.explained_variance_ratio_) #查看主成分比例

f1=plt.figure()
ax=f1.add_subplot(111)
X0=np.hstack([Y.reshape(-1,1),X_])
a=np.array([x for x in X0 if x[0]=='M'])
b=np.array([x for x in X0 if x[0]=='B'])
ax.scatter(a[:,1],a[:,2],marker='o',c='g')
ax.scatter(b[:,1],b[:,2],marker='*',c='m')
plt.show()       #显然X_分成了两簇，可以二分类
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

def I(b):
    if b==True:
        return 1
    else:
        return 0
def recall_rate(predict,primary):   #召回率为评价指标,召回率越大，模型效果越好
    count=0
    countf=0
    for i in range(predict.shape[0]):
        count+=I(predict[i]==primary[i]=='M')
        countf+=I(predict[i]=='B' and primary[i]=='M')
    return count/(count+countf)

model=LogisticRegression(multi_class='ovr')
X_train, X_test, Y_train, Y_test = train_test_split(X_, Y, test_size=0.3, random_state=42)
r=model.fit(X_train,Y_train)
y_train=r.predict(X_train)  #训练集的预测值
y_test=r.predict(X_test)   #测试集的预测值

print('训练误差:',recall_rate(y_train,Y_train))
print('泛化误差:',recall_rate(y_test,Y_test))
```

```python
import matplotlib.pyplot as plt
f2=plt.figure()
bx=f2.add_subplot(111)
bx.scatter(a[:,1],a[:,2],marker='o',c='g')
bx.scatter(b[:,1],b[:,2],marker='*',c='m')
w=r.coef_.reshape(-1,1)  #系数
b=r.intercept_  #截距
x=np.arange(-1000,1500,0.01).reshape(-1,1)
y=np.zeros_like(x)
y=(-b-w[0]*x)/w[1]  #分界直线方程
bx.plot(x,y,c='r')
plt.show()
```

```python
#降维前的分类
mod=LogisticRegression(multi_class='ovr')
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)  
p=mod.fit(X_train,Y_train)
y_train=p.predict(X_train)  #训练集的预测值
y_test=p.predict(X_test)   #测试集的预测值

print('训练误差：',recall_rate(y_train,Y_train))
print('泛化误差：',recall_rate(y_test,Y_test))  #降维前的召回率更低
```
