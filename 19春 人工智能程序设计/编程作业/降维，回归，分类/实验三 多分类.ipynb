```python
class my_pca(object):
    "å®šä¹‰my_pcaç±»ï¼Œä¼ å…¥çš„n*mçŸ©é˜µä¸ºnæ¡æ•°æ®mç»´"
    def __init__(self,X):
        mean=X.mean(axis=0)
        X=X-mean #ä¸­å¿ƒåŒ–åŽçš„çŸ©é˜µ
        self.X=X.T
        self.datapoints=self.X.shape[1]
    def cov(self):
        return np.dot(self.X,self.X.T)/self.datapoints  #è¿”å›žåæ–¹å·®çŸ©é˜µ
    def eigenvalues_vectors(self):
        x_cov=self.cov()
        a,b=np.linalg.eig(x_cov)  #ç‰¹å¾å€¼ä¸Žç‰¹å¾å‘é‡
        c=np.hstack([a.reshape(-1,1),b.T])  
        c_df=pd.DataFrame(c).sort_values(by=0,ascending=False)
        return c_df   
    def reduce_dim(self,n_components): #n_componentsä¸ºé™æˆçš„ç»´æ•° 
        self.n_components=n_components
        x_df=self.eigenvalues_vectors().values[:,1:]  #x_dfä¸ºä»¥ç‰¹å¾å€¼ä¸ºè¡Œå‘é‡çš„çŸ©é˜µ
        self.X_=np.dot(x_df,self.X).T
        return np.dot(x_df[:self.n_components,:],self.X).T
    def explained_variance(self): #ä¸»æˆåˆ†æ–¹å·®å€¼
        return np.var(self.reduce_dim(self.n_components),axis=0)
    def explained_variance_ratio(self):  #ä¸»æˆåˆ†æ–¹å·®æ¯”ä¾‹
        p=np.var(self.X_,axis=0)
        return (p/p.sum())[:self.n_components]
```
