```python
import pandas as pd
import numpy as np
columns=['class']+['feature'+str(x) for x in range(1,11)]
data=pd.read_csv('C:/Users/Jaqen/Desktop/dataset.csv',names=columns,index_col=False)
data.sort_values(by='class',inplace=True)
data
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

count=data.groupby('class').feature1.count()
count=[count[:i+1].sum() for i in range(len(count))]#储存类别分界处的Index
X=data.iloc[:,1:].values
Y=data['class'].values 
pca=PCA(n_components=3)
X_=pca.fit_transform(X) #降维
print(pca.explained_variance_ratio_) #查看主成分比例
print(X_)

f1=plt.figure()
ax=f1.add_subplot(111, projection='3d')
ax.scatter(X_[:count[0],0],X_[:count[0],1],X_[:count[0],2],marker='o',c='m')
ax.scatter(X_[count[0]:count[1],0],X_[count[0]:count[1],1],X_[count[0]:count[1],2],marker='o',c='y')
ax.scatter(X_[count[1]:count[2],0],X_[count[1]:count[2],1],X_[count[1]:count[2],2],marker='o',c='b')
ax.scatter(X_[count[2]:count[3],0],X_[count[2]:count[3],1],X_[count[2]:count[3],2],marker='o',c='c')
plt.show()  
```

```python
class my_pca(object):
    "定义my_pca类，传入的n*m矩阵为n条数据m维"
    def __init__(self,X):
        mean=X.mean(axis=0)
        X=X-mean #中心化后的矩阵
        self.X=X.T
        self.datapoints=self.X.shape[1]
    def cov(self):
        return np.dot(self.X,self.X.T)/self.datapoints  #返回协方差矩阵
    def eigenvalues_vectors(self):
        x_cov=self.cov()
        a,b=np.linalg.eig(x_cov)  #特征值与特征向量
        c=np.hstack([a.reshape(-1,1),b.T])  
        c_df=pd.DataFrame(c).sort_values(by=0,ascending=False)
        return c_df   
    def reduce_dim(self,n_components): #n_components为降成的维数 
        self.n_components=n_components
        x_df=self.eigenvalues_vectors().values[:,1:]  #x_df为以特征值为行向量的矩阵
        self.X_=np.dot(x_df,self.X).T
        return np.dot(x_df[:self.n_components,:],self.X).T
    def explained_variance(self): #主成分方差值
        return np.var(self.reduce_dim(self.n_components),axis=0)
    def explained_variance_ratio(self):  #主成分方差比例
        p=np.var(self.X_,axis=0)
        return (p/p.sum())[:self.n_components]
```

```python
#用自己写的PCA训练
my=my_pca(X)
my_X=my.reduce_dim(3)
print(my.explained_variance_ratio())
print(my_X)#可以看出两者降维后的矩阵除了有几列符号相反外其他都相同，而造成这个的原因是降维时选取的单位特征向量的符号不同

f2=plt.figure()
bx=f2.add_subplot(111, projection='3d')
bx.scatter(my_X[:count[0],0],my_X[:count[0],1],my_X[:count[0],2],marker='o',c='m')
bx.scatter(my_X[count[0]:count[1],0],my_X[count[0]:count[1],1],my_X[count[0]:count[1],2],marker='o',c='y')
bx.scatter(my_X[count[1]:count[2],0],my_X[count[1]:count[2],1],my_X[count[1]:count[2],2],marker='o',c='b')
bx.scatter(my_X[count[2]:count[3],0],my_X[count[2]:count[3],1],my_X[count[2]:count[3],2],marker='o',c='c')
plt.show()  
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

def I(b):
    if b==True:
        return 1
    else:
        return 0
def error_rate(predict,primary):   #错误率为评价指标
    count=0
    for i in range(predict.shape[0]):
        count+=I(predict[i]!=primary[i])  
    return count/predict.shape[0]

X_train, X_test, Y_train, Y_test = train_test_split(X_, Y, test_size=0.3, random_state=42)
lor=LogisticRegression(penalty='l1',C=100,multi_class='ovr') 
lor.fit(X_train,Y_train)
w=lor.coef_
print(lor.score(X_train,Y_train))
print(lor.score(X_test,Y_test))
y_train=lor.predict(X_train)
y_test=lor.predict(X_test)
print(error_rate(y_train,Y_train))
print(error_rate(y_test,Y_test))
```
