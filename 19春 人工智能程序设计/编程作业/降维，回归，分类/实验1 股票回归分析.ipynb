```python
import pandas as pd
import numpy as np
stocks=pd.read_csv('C:/Users/Jaqen/Desktop/data_akbilgic.csv')
stocks.columns=['date','ISE(TL-based)', 'ISE(usd)', 'SP', 'DAX', 'FTSE', 'NIKKEI', 'BOVESPA', 'EU', 'EM']
temp=stocks.drop('date',axis=1).drop(0).astype('float')#去掉data列与第0行，将剩余数据转成float
temp.insert(0,'date',stocks['date'])#第0列插入data数据
stocks=temp
stocks.reset_index(drop=True,inplace=True)
stocks
```

```python
import matplotlib.pyplot as plt
for i in range(2,6):
    plt.subplot(219+i)
    plt.scatter(stocks.iloc[:,i],stocks.iloc[:,1],marker='o',c='g',s=10)
plt.show()
for i in range(6,10):
    plt.subplot(215+i)
    plt.scatter(stocks.iloc[:,i],stocks.iloc[:,1],marker='o',c='g',s=10)
plt.show()    
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split

def mse(predict,primary):  #评估指标为差值均方
    return ((predict-primary)**2).sum()/predict.shape[0]

model=LinearRegression()
X=stocks.iloc[:,2:].values
Y=stocks.iloc[:,1].values.reshape(-1,1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)  #数据集划分
r=model.fit(X_train,Y_train)
y_test=r.predict(X_test)
y_train=r.predict(X_train)
print('训练误差：',mse(y_train,Y_train))  #训练误差
print('泛化误差：',mse(y_test,Y_test))    #泛化误差
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=4)
X_=pca.fit_transform(X) #降维
print(pca.explained_variance_ratio_) #查看主成分比例
X_train, X_test, Y_train, Y_test = train_test_split(X_, Y, test_size=0.3, random_state=42)  #数据集划分
r=model.fit(X_train,Y_train)
y_test=r.predict(X_test)
y_train=r.predict(X_train)
print('训练误差',mse(y_train,Y_train))  #训练误差
print('泛化误差',mse(y_test,Y_test))   #泛化误差
#降维后的误差稍大一些，随着主成分的增加趋近于原误差
```
