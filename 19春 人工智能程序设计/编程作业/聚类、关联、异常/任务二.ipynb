```python
class Apriori(object):
    def __init__(self,data,support,confidence):
        self.sup=support
        self.con=confidence
        self.data=data
        self.classes=[frozenset({x}) for x in set(data[:,-1])]
        
    def fit(self):
        data_lst=[]           #储存符合支持度的项集
        c=self.first_c()   #第一层，每个项集只有一个元素
        while len(c)>0:
            dict_c=self.scan(c)    #扫描，将不符合支持度的项集去掉
            if dict_c=={}:
                break              #如果都不符合则跳出
            data_lst.append(dict_c)   
            c=self.c_to_ck(list(dict_c.keys()))    #将字典的键取出，生成下一层  
        data_lst=dict([(i,x[i]) for x in data_lst for i in x.keys()])
        self.freq=data_lst
        return data_lst   
    
    def first_c(self):  #初始的单个项集
        c1=[]
        for t in self.data:
            for item in t:
                if [item] not in c1:
                    c1.append([item])
        c1.sort()
        c1=list(map(frozenset,c1))
        return c1 #形如 [frozenset({1}), frozenset({2}), frozenset({4})]
    
    def c_to_ck(self,c):  #从ck-1生成ck
        ck=[]
        k=len(c[0])  #每个项集的宽度
        if k==1:
            ck=[c[i]|c[j] for i in range(len(c)) for j in range(i+1,len(c))]  #直接求并集
            return ck
        ck=[c[i]|c[j] for i in range(len(c)) for j in range(i+1,len(c)) if list(c[i])[:k-1]==list(c[j])[:k-1]]
        return ck         #形如[frozenset({1, 2, 3}), frozenset({1, 2, 4})]
    
    def scan(self,ck):     #ck形如 [frozenset({1}), frozenset({2}), frozenset({4})]
        dic={}
        for can in ck:
            for t in self.data:
                if can.issubset(t):
                    dic[can]=dic.get(can,0)+1
        N=len(self.data)
        sup_dict={x:dic[x]/N for x in dic if dic[x]/N>=self.sup}
        return sup_dict #  返回一个字典，键为符合支持度的项集，值为支持度  形如{frozenset({1}): 1, frozenset({2}): 1}  
    
    def association(self):
        dict_a={}
        fre=self.freq
        fre_keys=list(fre.keys())
        for x in self.classes:
            if x in fre_keys:
                fre_keys.remove(x)
        for classes in self.classes:
            for item in fre_keys:
                if classes.issubset(item):
                    if fre.get(item)/fre.get(item-classes)>=self.con: 
                        dict_a[str(set(item-classes))+'-'+str(set(classes))]=fre[item]/fre[item-classes]
        return dict_a 
```

```python
ap=Apriori(data.values,0.15,0.6)   #查看哪些值与unacc关联
ap.fit()
ap.association()
```

```python
data['buying']=data['buying'].map(lambda x:-1 if x=='b-low' else 0 if x=='b-med' else 1 if x=='b-high' else 2 )
data['paint']=data['paint'].map(lambda x:-1 if x=='p-low' else 0 if x=='p-med' else 1 if x=='p-high' else 2 )
data['doors']=data['doors'].map(lambda x:2 if x=='d-2' else 3 if x=='d-3' else 4 if x=='d-4' else 5 )
data['persons']=data['persons'].map(lambda x:2 if x=='p-2' else 4 if x=='p-4' else 6)
data['lug_boot']=data['lug_boot'].map(lambda x:-1 if x=='l-small' else 0 if x=='l-med' else 1)
data['safety']=data['safety'].map(lambda x:-1 if x=='s-low' else 0 if x=='s-med' else 1)
data['class ']=data['class '].map(lambda x:0 if x=='unacc' else 1 if x=='acc' else 2 if x=='good' else 3 )
data
```

```python
from sklearn.neighbors import KNeighborsClassifier as KN
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler=StandardScaler()
X=data.values[:,:6]
X=scaler.fit_transform(X)
Y=data.values[:,6]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

knn=KN(weights='distance',n_neighbors=10)
knn.fit(X_train,Y_train)
print(knn.predict(X_test))
print(knn.score(X_test,Y_test))
```
